```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("train_0irEZ2H.csv")
df
```

```python
df.describe()
```

```python
df.hist(edgecolor="black",color="hotpink")
plt.tight_layout()
plt.plot()
```

```python
df.units_sold.hist(edgecolor="black",color="lightblue")
```

```python
df[["day","month","year"]]=df["week"].str.split("/",expand=True)
df=df.drop("week",axis =1)
df
```

```python
from sklearn.ensemble import  RandomForestRegressor
from sklearn.model_selection import train_test_split
X,y=df.drop("units_sold",axis=1),df["units_sold"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
model=RandomForestRegressor(n_jobs=-1)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error
y_pred=model.predict(X_test)
rmse=root_mean_squared_error(y_pred,y_test)
rmse
```

```python
plt.scatter(y_pred,y_test)
plt.plot(np.linspace(y_pred.min(),y_pred.max()),np.linspace(y_test.min(),y_test.max()),color="hotpink")
```

```python
model.score(X_test,y_test)
```
